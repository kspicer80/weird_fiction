```python
import numpy as np
import os
from nltk import word_tokenize
from nltk import sent_tokenize
from collections import Counter
import spacy 
nlp = spacy.load('en_core_web_lg')
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
```

```python
rb_tc = open('rb_tc.txt').read()
```

```python
rb_tc_nlp = nlp(rb_tc)
```

```python
len(rb_tc_nlp)
```

```python
rb_tc_sents = list(rb_tc_nlp.sents)
sentenceLengths = [len(sent) for sent in rb_tc_sents]
sentenceLengths[0]
```

```python
len(sentenceLengths)
```

```python
plt.plot(sentenceLengths)
plt.savefig('sentence_lengths_in_bradbury_the_crowd.png')
```
